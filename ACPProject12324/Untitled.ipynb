```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
path = "ex2_msd_0.dat"
path1 = "ex2_msd_20.dat"

t_0 = []
msd_0 = []
msad_0 = []

with open(path,'r') as f:
    for line in f:
        c = line.split()
        t_0.append(float(c[0]))
        msd_0.append(float(c[1]))
        msad_0.append(float(c[2]))
        
t_20 = []
msd_20 = []
msad_20 = []

with open(path1,'r') as f:
    for line in f:
        c = line.split()
        t_20.append(float(c[0]))
        msd_20.append(float(c[1]))
        msad_20.append(float(c[2]))
```

```python
plt.plot(t_0,msd_0,color='r',label='Pe=0')
plt.plot(t_20,msd_20,color='b',label='Pe=20')

plt.xlabel('t')
plt.ylabel('MSD')
plt.xscale('log')
plt.yscale('log')
plt.legend()
```
