```python
#1f benchmark

path = 'ex1f_runtime.dat'

N = []
t_naive = []
t_cell = []

with open(path,'r') as f:
    for line in f:
        columns = line.split()
        N.append(int(columns[0]))
        t_naive.append(float(columns[1]))
        t_cell.append(float(columns[2]))
        
```

```python
fig = plt.figure(figsize = (10.8,9))

coef_naive = np.polyfit(N,t_naive,2)
print(coef_naive)
coef_cell = np.polyfit(N,t_cell,1)
print(coef_cell)



def O_N(x,a,b):
    return x*a + b
def O_N2(x,a,b,c):
    return a*x**2 + b* x + c
    
x = np.arange(N[0],N[-1],0.1)

plt.plot(x,O_N2(x,coef_naive[0],coef_naive[1],coef_naive[2]),color='deepskyblue',label=r'$f_2(N)= a \cdot N^2 + b \cdot N +  c$')
plt.plot(x,O_N(x,coef_cell[0],coef_cell[1]),color='orangered',label=r'$f_1(N)= d \cdot N + e$')    
    
plt.scatter(N, t_naive,color='blue',label='Naive algorithm')
plt.scatter(N, t_cell,color='red',label='Cell list algorithm')

plt.xlabel('Number of points N',fontsize=14)
plt.ylabel('Computation time t $[\mu s]$',fontsize=14)
plt.xscale('log')
plt.yscale('log')

plt.legend()

plt.xlim(5,2*10**5)
plt.ylim(5,10**8)

plt.savefig('1f_runtime.png',dpi=400)
```
